```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.metrics import roc_auc_score
from xgboost import XGBClassifier

# Load data
data = pd.read_csv(r"C:\Users\ankit\datathon\dataset and all\training_set_features.csv")
dataset = pd.read_csv(r"C:\Users\ankit\datathon\dataset and all\training_set_labels.csv")

# Merge datasets
data = pd.merge(data, dataset, on='respondent_id', how='inner')

# Check for any missing values
print(data.isnull().sum())


```

```python
numeric_features = data.select_dtypes(include=['int64', 'float64']).columns
categorical_features = data.select_dtypes(include=['object']).columns


imputer = SimpleImputer(strategy='most_frequent')
data[numeric_features] = imputer.fit_transform(data[numeric_features])
```

```python
label_encoders = {}
for col in categorical_features:
    label_encoders[col] = LabelEncoder()
    data[col] = label_encoders[col].fit_transform(data[col])
```

```python
X = data.drop(['xyz_vaccine', 'seasonal_vaccine'], axis=1)  # Ensure respondent_id is not included here
y1= data[['xyz_vaccine']]
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_curve, roc_auc_score

# Assuming X and y are your features and target after some preprocessing
X_train, X_test, y_train, y_test = train_test_split(X, y1, test_size=0.2, random_state=42)

# Define Random Forest model
model_rf = RandomForestClassifier(bootstrap=True,
    min_samples_leaf=1,
    min_samples_split=10,                           
    n_estimators=200,
    max_depth=10,
    random_state=100
)

# Fit Random Forest model
model_rf.fit(X_train, y_train)

# Predict probabilities
pred_prob_rf = model_rf.predict_proba(X_test)[:, 1]

# Calculate ROC curve and AUC score
fpr_rf, tpr_rf, _ = roc_curve(y_test, pred_prob_rf, pos_label=1)
auc_rf = roc_auc_score(y_test, pred_prob_rf)

# Print AUC score for Random Forest
print(f"AUC Score (Random Forest): {auc_rf}")
```
